```python
import pandas as pd
import numpy as np
```

```python
# Importing CSV

data = pd.read_csv('iris-data.csv')
data.head()
```

```python
# Checking if the types of the columns are correct

data.dtypes
```

```python
# Changing name of the column 'class' because class is already something on Python to prevent any mistakes

data.rename(columns={'class': 'species'}, inplace=True)
```

```python
# Finding mistakes in the 'class' column

data.groupby(by='species').count()
```

```python
# Replacing 'Iris-setossa' mistake for 'Iris-setosa'

data['species'] = data['species'].str.replace('Iris-setossa', 'Iris-setosa')
set(data['species'])
```

```python
# Replacing 'versicolor' mistake for 'Iris-versicolor'

data['species'] = data['species'].str.replace('versicolor', 'Iris-versicolor')
set(data['species'])
```

```python
data['species'] = data['species'].str.replace('Iris-Iris-versicolor', 'Iris-versicolor')
set(data['species'])
```

```python
# Finding rows with NaN values

data[data.isnull().any(axis=1)]
```

```python
# We know from grouping by class before that there's 50 rows of Iris-setosa, and only 5 NaN, so instead
# of dropping the NaN values we will replace them with the mean of the series for that class.

#Checking the mean of petal_width_cm for Iris-setosa
data[data['species']=='Iris-setosa'].mean()
```

```python
# Filling NaNs with the mean

data = data.fillna(0.25)
```

```python
# Checking that there is no NaNs left

data[data.isnull().any(axis=1)]
```
